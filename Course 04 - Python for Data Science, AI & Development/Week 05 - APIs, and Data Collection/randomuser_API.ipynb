```python
## import the package
from randomuser import RandomUser
## import pandas to make dataframe which is fetch out of API
import pandas as pd
```

```python
## initiale the API element
r = RandomUser()
help(r)
```

```python
## static method is generate_users()
## it will generate randomly selected users.
some_users = r.generate_users(20)
#some_users
```

```python
## now i need to get more details about the user like full name, DOB, city, email, etc.
for user in some_users:
    print(user.get_full_name(), "--", user.get_email())
```

```python
## putting everything into a dataframe
def get_user_details_into_dataframe():
    users = []
    
    for user in RandomUser.generate_users(25):
        users.append({"Name" : user.get_full_name(),
                         "Gender" : user.get_gender(),
                         "City" : user.get_city(),
                         "State" : user.get_state(),
                         "Email" : user.get_email(),
                         "DOB" : user.get_dob,
                         "Picture" : user.get_picture()})
    return pd.DataFrame(users)
```

```python
get_user_details_into_dataframe()
```
